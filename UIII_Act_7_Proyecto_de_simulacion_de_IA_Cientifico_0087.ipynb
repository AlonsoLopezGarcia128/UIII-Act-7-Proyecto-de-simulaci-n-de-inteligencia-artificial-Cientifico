### Simulacion de inteligencia artificial **Cientifico y Laboratorio**

```python
# SIMULADOR DE IA PARA LABORATORIO

print("🔬 INICIANDO SISTEMA DE IA: LAB-BOT 🔬")
nombre = input("Ingresa tu nombre, científico: ")

print("Bienvenido al laboratorio,", nombre)

opcion = ""

# Ciclo principal
while opcion != "4":
    print("\n--- MENÚ PRINCIPAL ---")
    print("1. Analizar sustancia")
    print("2. Calcular temperatura")
    print("3. Diagnóstico del experimento")
    print("4. Salir")

    opcion = input("Selecciona una opción: ")

    # Opción 1
    if opcion == "1":
        sustancia = input("Ingresa el tipo de sustancia (acido/base/neutra): ")

        if sustancia == "acido":
            print("⚠️ La sustancia es corrosiva. Manejar con cuidado.")
        elif sustancia == "base":
            print("🧪 La sustancia puede reaccionar con ácidos.")
        elif sustancia == "neutra":
            print("✅ Sustancia estable, sin riesgos.")
        else:
            print("❌ Sustancia desconocida.")

    # Opción 2
    elif opcion == "2":
        temp = int(input("Ingresa la temperatura del experimento: "))

        if temp < 0:
            print("❄️ Temperatura muy baja. Riesgo de congelación.")
        elif temp >= 0 and temp <= 100:
            print("🌡️ Temperatura adecuada para el experimento.")
        elif temp > 100:
            print("🔥 Temperatura alta. Precaución.")
        else:
            print("❌ Error en temperatura.")

    # Opción 3
    elif opcion == "3":
        resultado = input("Resultado del experimento (exitoso/fallido): ")

        if resultado == "exitoso":
            print("🎉 Experimento completado con éxito.")
        elif resultado == "fallido":
            print("🔁 Revisa variables y vuelve a intentar.")
        else:
            print("❌ Resultado no válido.")

    # Opción 4
    elif opcion == "4":
        print("👋 Cerrando sistema de laboratorio...")

    # Opción inválida
    else:
        print("❌ Opción no válida. Intenta de nuevo.")

print("Sistema apagado.")
print("Programa realizado por Alonso Lopez NL = 31")
```
